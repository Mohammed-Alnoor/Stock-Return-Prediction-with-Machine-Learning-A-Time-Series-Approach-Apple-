# Apple (AAPL) - Feature Engineering

## Objective
Transform raw stock data into meaningful features that can be used for predictive modeling.

## Key Idea
Raw price data is not suitable for machine learning models due to non-stationarity and noise.
We will create derived features that capture trends, momentum, and volatility.

## Load and Prepare Data



```python
import pandas as pd
import numpy as np

stock = pd.read_csv('../data/raw/aapl_stock_prices.csv')

stock['Date'] = pd.to_datetime(stock['Date'])
stock = stock.sort_values('Date')
stock.set_index('Date', inplace=True)
```

## Returns Calculation

We compute daily returns using percentage change.


```python
stock['Return'] = stock['Close'].pct_change()
```

## Lag Features

Lag features represent past values of the data.

```python
stock['Close_lag_1'] = stock['Close'].shift(1)
stock['Close_lag_2'] = stock['Close'].shift(2)
stock['Close_lag_3'] = stock['Close'].shift(3)

stock['Return_lag_1'] = stock['Return'].shift(1)
stock['Return_lag_2'] = stock['Return'].shift(2)
```

## Moving Averages

Moving averages smooth out price fluctuations and highlight trends.


```python
stock['MA_7'] = stock['Close'].rolling(7).mean()
stock['MA_30'] = stock['Close'].rolling(30).mean()
stock['MA_50'] = stock['Close'].rolling(50).mean()
```

## Volatility Features

Volatility measures how much the price fluctuates over time.



```python
stock['Volatility_7'] = stock['Return'].rolling(7).std()
stock['Volatility_30'] = stock['Return'].rolling(30).std()
```

## Volume Features

Volume represents trading activity.


```python
stock['Volume_change'] = stock['Volume'].pct_change()
stock['Volume_MA_7'] = stock['Volume'].rolling(7).mean()
```

## Price-Based Features

We create features from price differences.

```python
stock['High_Low_diff'] = stock['High'] - stock['Low']
stock['Open_Close_diff'] = stock['Open'] - stock['Close']
```

## Target Variable

We define the prediction target as the next day's return.


```python
stock['Target'] = stock['Return'].shift(-1)
```

```python
stock['MA_20'] = stock['Close'].rolling(20).mean()
stock['MA_50'] = stock['Close'].rolling(50).mean()

stock['MA_ratio'] = stock['MA_20'] / stock['MA_50']
```

```python
stock['Momentum_5'] = stock['Close'] - stock['Close'].shift(5)
stock['Momentum_10'] = stock['Close'] - stock['Close'].shift(10)
```

```python
stock['Volatility_14'] = stock['Return'].rolling(14).std()
```

```python
stock['Close_to_High'] = stock['Close'] / stock['High']
stock['Close_to_Low'] = stock['Close'] / stock['Low']
```

```python
stock = stock.replace([np.inf, -np.inf], np.nan)
stock = stock.dropna()
```

## Data Cleaning

After creating features:
- Remove missing values caused by rolling and shifting

```python
stock = stock.dropna()
```

```python
stock.reset_index().to_csv('../data/processed/feature_data.csv', index=False)
```

## Feature Engineering Summary

- Created lag features to capture temporal dependencies
- Generated moving averages to represent trends
- Calculated volatility to measure market risk
- Built volume-based features for trading activity
- Defined target as next-day return for prediction
